```python
# Stacking and Splitting
s = np.array([[1,2],[3,4],[11,12],[21,22]])
print(s)

print("\n" + "*" * 15 + "\n") # Cosmetics

t = np.array([[5,6],[7,8],[15,16],[25,26]])
print(t)

print("\n" + "*" * 15 + "\n") # Cosmetics

u = np.vstack((s,t))
print("vstack : " + "\n", u)

print("\n" + "*" * 15 + "\n") # Cosmetics

v = np.hstack((s,t))
print("hstack : " + "\n", v)

print("\n" + "*" * 15 + "\n") # Cosmetics

w = np.split(s, 2)
print("split : " + "\n", w)

print("\n" + "*" * 15 + "\n") # Cosmetics

x = np.vsplit(s, 4)
print("vsplit : " + "\n", x)

print("\n" + "*" * 15 + "\n") # Cosmetics

y = np.hsplit(t, 2)
print("hsplit : " + "\n", y)
```

```python
# Matrix opertions
h = np.array([[[1, 2], [3, 4]],[[11, 12], [13, 14]]])
print("h : " + "\n", h)

print("\n" + "*" * 15 + "\n") # Cosmetics

i = np.array([[[5, 6], [7, 8]],[[15, 16], [17, 18]]])
print("i : " + "\n", i)

print("\n" + "*" * 15 + "\n") # Cosmetics

j = np.dot(h, i)   # matrix multiplication
print("Dot product : " + "\n", j)

print("\n" + "*" * 15 + "\n") # Cosmetics

k = np.cross(h, i)
print("Cross product : " + "\n", k)

print("\n" + "*" * 15 + "\n") # Cosmetics

l = np.linalg.inv(h)
print("Inverse : " + "\n", l)

print("\n" + "*" * 15 + "\n") # Cosmetics

m = np.linalg.det(h)
print("Determinant : " + "\n", m)
```

```python
# Linear Algebra
nn = np.array([[1, 2], [3, 4]])
mm = np.array([[0, -3], [0.5, -0.1]])
eigenvalues, eigenvectors = np.linalg.eig(nn)

print("Eigenvalues:" + "\n", eigenvalues)

print("\n" + "*" * 15 + "\n") # Cosmetics

print("Eigenvectors:" + "\n", eigenvectors)

print("\n" + "*" * 15 + "\n") # Cosmetics

correlation_coefficient = np.corrcoef(nn,mm)

print("Correlation Coefficient:" + "\n", correlation_coefficient)
# correlation_coefficient[0, 1] represents the correlation between the first column of n and the second column of m
```

```python
# Normalize an array
z = np.array([1, 2, 3, 4, 5])
normalized_array = z / np.linalg.norm(z)
print("Normalized array:" +"\n", normalized_array)
```

```python
# Solve linear equation

# 2x+3y=8 & 5x−2y=−7
a = np.array([[2, 3], [5, -2]]) # Define coefficients and constants
b = np.array([8, -7])

solution = np.linalg.solve(a, b) # Solve the linear equation
print("Solution:" +"\n", solution)

x = solution[0]
y = solution[1]

print(f"x = {x}, y = {y}")
```

```python
# Sdandarize a array
z = np.random.randint(1,100,25)
print(z)

# Calculate mean and standard deviation
mean = np.mean(z)
std_dev = np.std(z)

# Standardize the array
standardized_arr = (z - mean) / std_dev

print("\nMean:", mean)
print("Standard Deviation:", std_dev)
print("\nStandardized Array:")
print(standardized_arr)
```

```python
# Bin Count

arr = np.array([1, 2, 2, 3, 3, 3, 4, 4, 4, 4,5]) # Create an array of non-negative integers

# Calculate bin counts
bin_counts = np.bincount(arr)

print("Original Array:", arr)
print("Bin Counts:", bin_counts)
```

```python
#Iterating Over Array
a = np.arange(0,60,5).reshape(3,4)
print(a)

for x in np.nditer(a): # It is an efficient multidimensional iterator object using which helps to iterate over an array.
   print(x,end=" ")
```

```python
# Broadcasting

x = np.array([[1], [2], [3]]) 
y = np.array([4, 5, 6])  
   
# to broadcast x against y 
b = np.broadcast(x,y)  
# it has an iterator property, a tuple of iterators along self's "components." 

for a in b:
    print(a,end=" ")
```
