```python
# Hot Deck Imputation
import random

def hot_deck_imputation(row):
    if pd.isnull(row['Type 2']):
        similar_rows = pokemon_df[pokemon_df['Type 2'].notnull()]
        random_similar_row = similar_rows.sample(1)
        return random_similar_row['Type 2'].values[0]
    else:
        return row['Type 2']

pokemon_df['Type 2'] = pokemon_df.apply(hot_deck_imputation, axis=1)
pokemon_df['Type 2'].value_counts()
```

## Handling outliers

```python
# Z score
from scipy import stats

z_scores_math = stats.zscore(df['Math'])
z_scores_science = stats.zscore(df['Science'])

outliers_math = (z_scores_math > 2) | (z_scores_math < -2)
outliers_science = (z_scores_science > 2) | (z_scores_science < -2)

print(outliers_math)
print(outliers_science)

rows_to_remove = outliers_math | outliers_science

df_cleaned = df[~rows_to_remove]

print('\n', df_cleaned)
```

```python
# IQR

Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

# Identify outliers
outliers = ((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR)))

# Remove outliers
df_iqr_cleaned = df[~outliers.any(axis=1)]

print(df_iqr_cleaned)
```

```python
# Capping and flooring

lower_threshold = 10  # Example threshold, you may choose based on your data
upper_threshold = 100

# Select numeric columns for capping and flooring
numeric_columns = df.select_dtypes(include=['number'])

# Cap and floor the data
df_capped_floored = numeric_columns.clip(lower=lower_threshold, upper=upper_threshold)

# Combine with non-numeric columns
df_cleaned = pd.concat([df['Name'], df_capped_floored], axis=1)

print(df_cleaned)
```

```python
# winsorizing: Replace the extreme values with values closer to the mean or median
from scipy.stats.mstats import winsorize

# Apply winsorizing to numeric columns
numeric_columns = df.select_dtypes(include=['number'])
winsorized_data = numeric_columns.apply(lambda x: winsorize(x, limits=[0.05, 0.05]))

# Create a DataFrame with winsorized values
df_winsorized = pd.concat([df['Name'], winsorized_data], axis=1)

print(df_winsorized)
```
