#### Working with Multiple Dataframes

- Appending (Stacking Vertically)
- Concatenation (Stacking Horizontally)
- SQL style Merging

```python
# Appending
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.append.html    
```

```python
num_list = []
sq_list = []
for i in range(1, 11):
    num_list.append(i)
    sq_list.append(i*i)

sq_dict = dict(zip(num_list, sq_list))
sq_dict
```

```python
sq_df = pd.DataFrame(data = sq_dict.values(), index = sq_dict.keys(), columns = ['Square'])
sq_df
```

```python
new_list = []
new_sq = []
for i in range(11, 16):
    new_list.append(i)
    new_sq.append(i*i)

new_sq_dict = dict(zip(new_list, new_sq))
new_sq_df = pd.DataFrame(data = new_sq_dict.values(), index = new_sq_dict.keys(), columns = ['Square'])
new_sq_df
```

```python
sq_df.append(new_sq_df)
```

```python
sq_df.append(new_sq_df, ignore_index=True)
```

```python
# Concatenation

# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.concat.html
```

```python
sq_df = pd.DataFrame(data = sq_dict.values(), index = sq_dict.keys(), columns = ['Square'])
sq_df
```

```python
new_list = []
new_cubes = []
for i in range(6, 16):
    new_list.append(i)
    new_cubes.append(i**3)

cubes_df = pd.DataFrame(data = new_cubes, index = new_list, columns = ['Cube'])
cubes_df
```

```python
sq_cubes_df = pd.concat([sq_df, cubes_df], axis = 1)
sq_cubes_df
```

```python
# If axis = 0 or not given, it will simply stack one on top of the other - like append()
```

```python
pd.concat([sq_df, cubes_df])
```

```python
pd.concat([sq_df, cubes_df], axis = 1)
```

```python
pd.concat([sq_df, cubes_df], axis = 1, join = 'inner')
```

#### Applying Functions in Pandas

```python
def odd_even(n):
    if n % 2 == 0:
        return 'Even'
    else:
        return 'Odd'
```

```python
odd_even(23)
```

```python
odd_even(26)
```

```python
inner_df2['Attack Odd or Even'] = inner_df2['Attack : left'].apply(odd_even)
inner_df2
```

```python
inner_df2['Attack Odd or Even Lambda'] = inner_df2['Attack : left'].apply(lambda x : "Even" if x %2 == 0 else "Odd")
inner_df2
```

```python
def begins_with_vowel(input_str):
    if input_str.upper()[0] in ['A', 'E', 'I', 'O', 'U']:
        return True
    else:
        return False
```

```python
begins_with_vowel('Vivek')
```

```python
pokemon_df['Begins with Vowel'] = pokemon_df['Name'].apply(begins_with_vowel)
pokemon_df
```

```python
pokemon_df['Begins with Vowel'].value_counts()
```
